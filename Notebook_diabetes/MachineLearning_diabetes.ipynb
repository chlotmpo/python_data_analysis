## Machine Learning

### Importation of all the librairies that we will use

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
```

### Importation of the dataset

```python
diabetes_df = pd.read_csv(r'../Dataset/diabetic_data.csv', sep =',', na_values="?", low_memory=False)
diabetes_df
```

```python
diabetes_df.info()
```

### Transformation of categorical data in numeric data

```python
label_encoder = LabelEncoder()

for column in diabetes_df.select_dtypes(include=['object']).columns:
    diabetes_df[column] = label_encoder.fit_transform(diabetes_df[column])
```

```python
diabetes_df
```

```python
#df = pd.get_dummies(diabetes_df)
```

```python
from sklearn.model_selection import train_test_split

x = diabetes_df.loc[:, diabetes_df.columns != 'readmitted']
y = diabetes_df.loc[:, 'readmitted']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.33)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

###  K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 100)
knn.fit(x_train, y_train)
```

```python
#from sklearn.neighbors import KNeighborsClassifier
#classifier = KNeighborsClassifier(n_neighbors = 1000, metric = 'minkowski', p = 2)
#classifier.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
ac = accuracy_score(y_test,y_pred)
```

```python
ks = [x for x in np.arange(10,102, 10)]
acs = []

for k in ks:
    print(k)
    knn = KNeighborsClassifier(n_neighbors = k).fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    ac = accuracy_score(y_test,y_pred)
    acs.append(ac)
    
plt.plot(ks,acs,color = 'blue',linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
```
